```python
#step-3: handling missing data
#scenario 1: missing value in numerical column
#scenario 2: missing values in non numerical column
#scenario 3: having missing values in huge data sets
import numpy as np
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(X[:,1:3])
X[:,1:3]=imputer.transform(X[:,1:3])
```

```python
X
```

```python
#step 4: handling the catogoroal data
#which converts the data into numerica data
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
#Alexa-0,0,1
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
#encoder is a method which converts text to numerical value
X=ct.fit_transform(X)
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Y=le.fit_transform(Y)
```

```python
Y
```

```python
#step-5: split the data into training and testing data
from sklearn.model_selection import train_test_split#train test is a class which was imported froms model_selection ,used for spliting the data into training and testing
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
#step 6:feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train[:,3:]=sc.fit_transform(X_train[:,3:])
X_test[:,3:]=sc.transform(X_test[:,3:])


```
